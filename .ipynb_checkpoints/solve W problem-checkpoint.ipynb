```python
#just E and I
#find symbolic firing rates
from sympy import Matrix, solve_linear_system
from sympy.abc import a,b,d,e,j,k,x,y

#a:=jee
#b:=jei
#d:=jie
#e:=jii
#j:=je0
#k:=ji0
W = Matrix([[a,b,j],[d,e,k]])

#x:=re, y:=ri
sol=solve_linear_system(W,x,y)
print sol
```

```python
#E, I: balanced solution with parameters from sompolinsky paper 
#use to compare to simulated firing rates
from sympy import Matrix, solve_linear_system
from sympy.abc import a,b,c,d,e,f,g,h,i,j,k,l,m,w,x,y,z
from numpy import *

K = 1000.0 #average number of E->E connections per neuron

CEM = 1.0 #(μF/cm^2)
CIM = 1.0

gEL = 0.05 #(mS/cm^2)
gIL = 0.05

taue = CEM/gEL #membrane time constant for exc. neurons (ms)
taui = CIM/gIL
taua = 100
tau1 = 3
tau2 = 1

VL = -65 #(mV) leak potential
Vth = -50 #spiking threshold
VE = 0 #excitatory reversal potential
VI = -80 #inhibitory RP
Vde = VE-VL
Vdi = VI-VL

Ne = 20000.0
Ni = 5000.0
N0 = 20000.0
Ns=Ne+Ni+N0
p=Ne/Ns
q=Ni/Ns
r=N0/Ns

pe0 = K/N0
pi0 = pe0
pii = K/Ni
pei = pii
pee = K/Ne
pie = pee

sqrtK=sqrt(K)

je0 = pe0*r*4.846*Vde*gEL*tau1/sqrtK
jp0 = pi0*r*3.808*Vde*gIL*tau1/sqrtK
jee = pee*p*0.462*Vde*gEL*tau1/sqrtK
jpe = pie*p*1.270*Vde*gIL*tau1/sqrtK
jep = pei*q*10.5*Vdi*gEL*tau1/sqrtK
jpp = pii*q*9.5*Vdi*gIL*tau1/sqrtK

W = Matrix([[jee,jep,je0,0],[jpe,jpp,jp0,0]])
z=10.18
sol=solve_linear_system(W,x,y,z)
print sol
```

```python
#(E, PV, SOM)(L,R) with ratios of connection probabilities
#find symbolic firing rates

from sympy import Matrix, solve_linear_system
from sympy.abc import a,b,d,e,f,g,h,i,j,k,l,m,u,v,w,x,y,z

#a:=jee
#b:=jep
#e:=jpe
#f:=jpp
#k:=je0
#l:=jp0
#h := reps, ratio of jes:jep (excitatory connection from SOM:from PV)
#g := rrps, ratio of jsp:jps (reciprocal connections between i pops from PV:from SOM)
#j := rpse, ratio of jse:jpe (excitatory connection to SOM:to PV)
#i := rips, ratio of jss:jpp (interconnection within SOM:within PV)
#d := rps0, ratio of js0:jp0 (input layer to SOM:to PV)

W = Matrix([[a,a,b,0,h*b,0,k],[a,a,0,b,0,h*b,k],[e,e,f,0,g*f,0,d*k],[e,e,0,f,0,g*f,d*k],[j*e,j*e,f,0,i*f,0,d*k],[j*e,j*e,0,f,0,i*f,d*k]])

sol=solve_linear_system(W,u,v,w,x,y,z)
print sol
```

```python
#E, PV, SOM and left and right: balanced solution
from sympy import Matrix, solve_linear_system
from sympy.abc import a,b,c,d,e,f,t,u,v,w,x,y,z
from numpy import *

K = 1000.0 #average number of E->E connections per neuron

##tau and g are inversely related UNLESS THIS IS NOT 1
CEM = 1.0 #(μF/cm^2)
CIM = 1.0


#if you change gEL, you also need to change it in sompsim!
gEL = 0.05 #(mS/cm^2)
gIL = 0.05

taue = CEM/gEL #membrane time constant for exc. neurons (ms)
taui = CIM/gIL
taua = 100
tau1 = 3
tau2 = 1

VL = -65 #(mV) leak potential
Vth = -50 #spiking threshold
VE = 0 #excitatory reversal potential
VI = -80 #inhibitory RP
Vde = VE-VL
Vdi = VI-VL

Nel = 10000.
Ner = 10000.
Ne=Nel+Ner
Npl = 1250.
Npr = 1250.
Np=Npl+Npr
Nsl = 1250.
Nsr = 1250.
Ns=Nsl+Nsr
Ni=Np+Ns
N0 = 20000.

N=Ne+Np+Ns+N0
p=Nel/N
q=Npl/N
rl=Nsl/N
rr=Nsr/N
s=N0/N

rpse=1.4 #ratio of jse:jpe (excitatory connection to SOM:to PV)
reps=1.2 #ratio of jes:jep (excitatory connection from SOM:from PV)
rips=2. #ratio of jss:jpp (interconnection within SOM:within PV)
rrps=.7 #ratio of jsp:jps (reciprocal connections between i pops from PV:from SOM)
rps0=1. #ratio of js0:jp0 (input layer to SOM:to PV)
delta=0

sqrtK = sqrt(K)

#connection probabilities
pe0 = K/N0
pp0 = pe0
ps0 = pp0*rps0
ppp = K/Np
pep = ppp
pes = pep*reps
pps = K/Ns
psp = pps*rrps
pss = ppp*rips
pee = K/Ne
ppe = pee
pse = ppe*rpse


je0 = pe0*s*4.846*Vde*gEL*tau1/sqrtK
jp0 = pp0*s*3.808*Vde*gIL*tau1/sqrtK
js0 = ps0*s*3.808*Vde*gIL*tau1/sqrtK
jee = pee*p*0.462*Vde*gEL*tau1/sqrtK
jpe = ppe*p*1.270*Vde*gIL*tau1/sqrtK
jse = pse*p*Vde*gIL*tau1/sqrtK
jep = pep*q*10.5*Vdi*gEL*tau1/sqrtK
jpp = ppp*q*9.5*Vdi*gIL*tau1/sqrtK
jpsl = pps*rl*9.5*Vdi*gIL*tau1/sqrtK
jpsr = pps*rr*9.5*Vdi*gIL*tau1/sqrtK
jsp = psp*q*9.5*Vdi*gIL*tau1/sqrtK
jesl = pes*rl*10.5*Vdi*gEL*tau1/sqrtK
jesr = pes*rr*10.5*Vdi*gEL*tau1/sqrtK
jssl = pss*rl*9.5*Vdi*gIL*tau1/sqrtK
jssr = pss*rr*9.5*Vdi*gIL*tau1/sqrtK

W = Matrix([[jee,jee,jep,0,jesl,0,je0,0],[jee,jee,0,jep,0,jesr,je0,0],[jpe,jpe,jpp,0,jpsl,0,jp0,0],[jpe,jpe,0,jpp,0,jpsr,jp0,0],[jse,jse,jsp,0,jssl,0,js0,0],[jse,jse,0,jsp,0,jssr,js0,0]])
#print W
sol=solve_linear_system(W,t,u,v,w,x,y,z)
print sol

```
